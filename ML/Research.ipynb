```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV


data = pd.read_csv('Customers.csv')
data.dropna(inplace=True)

X_data = data[['Age', 'Salary']]
X_data = (X_data - X_data.mean()) / X_data.std()
Y_data = data['Purchased']
```

```python
data.info()
```

```python
print(sum(Y_data == 0), sum(Y_data == 1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

param_grid = {
    'n_neighbors': np.arange(1, 25), 
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']}

knn_gscv = GridSearchCV(
    model, 
    param_grid, 
    cv=5,
    scoring='accuracy',
    n_jobs=-1)
knn_gscv.fit(X_data, Y_data)
```

```python
pd.DataFrame(knn_gscv.cv_results_).sort_values('mean_test_score', ascending=False).head(10)
```

```python
from sklearn.svm import SVC
model = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
}

svm_gscv = GridSearchCV(
    model, 
    param_grid, 
    cv=5, 
    scoring='accuracy',
    n_jobs=-1)
svm_gscv.fit(X_data, Y_data)
```

```python
pd.DataFrame(svm_gscv.cv_results_).sort_values('mean_test_score', ascending=False).head(10)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

model = SVC()

param_grid = {
    'C': uniform(0.01, 10),
    'gamma': uniform(0.01, 10),
    'kernel': ['rbf']
}

svm_rscv = RandomizedSearchCV(
    model, 
    param_grid, 
    cv=5,
    scoring='accuracy',
    n_iter=5000,
    n_jobs=-1)
svm_rscv.fit(X_data, Y_data)
```

```python
pd.DataFrame(svm_rscv.cv_results_).sort_values('mean_test_score', ascending=False).head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

param_grid = {
    'n_estimators': [10, 30, 70, 100],
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 3, 4],
    'min_samples_split': [5, 7, 9],
    'min_samples_leaf': [6, 8, 10, 12]
}

rf_gscv = GridSearchCV(model, param_grid, cv=5, n_jobs=-1)
rf_gscv.fit(X_data, Y_data)
```

```python
pd.DataFrame(rf_gscv.cv_results_).sort_values('mean_test_score', ascending=False).head(10)
```

```python
model = RandomForestClassifier()

param_grid = {
    'n_estimators': range(10, 151),
    'criterion': ['gini', 'entropy'],
    'max_depth': range(2, 17),
    'min_samples_split': range(4, 21),
    'min_samples_leaf': range(4, 21)
}

rf_rscv = RandomizedSearchCV(
    model, 
    param_grid, 
    cv=5, 
    n_iter=2500, 
    n_jobs=-1)
rf_rscv.fit(X_data, Y_data)
```

```python
pd.DataFrame(rf_rscv.cv_results_).sort_values('mean_test_score', ascending=False).head(10)
```
